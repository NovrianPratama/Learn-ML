```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# baca file dataset
card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# Cek distribusi dataset khusus untuk label atau target
card_data['Class'].value_counts()
```

0 - Legit Transaction

1 - Fraudulent Transaction

```python
# Membagi dataset legit and fraudelent
legit = card_data[card_data.Class == 0]
fraud = card_data[card_data.Class == 1]
```

```python
print(legit.shape, fraud.shape)
```

Hasilnya menunjukkan perbandingan jumlah dataset yang sangat jauh dan tdiak seimbang

Mengubah jjumlah datasets nya menjadi seimbang.

Legit & Fraud -> 490

```python
# mengubah jumlah dataset untuk label menjadi 490 orang
legit_sample = legit.sample(n=490)
```

```python
legit_sample.shape
```

```python
# Concantenate 2 dataframe (Menggabungkan dua dataset, tpi khusus untuk row)
# axis=0 yaitu baris, jadi saya ingin concate pada baris, jika axis=1
# itu akan menambahkan kolom
# untuk legitsample dan fraud disimpan dalam list [], karena kedua variabel
## tersebut adalah dataframe

new_dataset = pd.concat([legit_sample, fraud], axis=0)
```
