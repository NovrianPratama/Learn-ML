# **Label Encoding**
- Convert the label into numeric form

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('/content/cancer.csv')
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

Label Encoding dataset brest cancer

```python
data.head()
```

```python
data.shape
```

```python
data['diagnosis'].value_counts()

# B mendiagnosiskan mulai terkena kanker
# M diagnosis sudah parah terkenan kanker
```

```python
# membuat variabel baru untuk label encoder
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(data.diagnosis)

# Penjelasan kode:
# fit(): Mempelajari label yang ada dalam kolom data.diagnosis
# transform(): Mengubah nilai kategori dalam kolom data.diagnosis menjadi
# angka berdasarkan kategori yang telah dipelajari oleh fit().
```

```python
# appending/menggabungkan label ke dataframe cancer
data['target'] = labels
```

# Lebel encoding iris dataset

```python
# import dataset iris
iris_df = pd.read_csv('/content/iris_data.csv')
```

```python
iris_df.head()
```

```python
iris_df.shape
```

```python
iris_df['Species'].value_counts()
```

```python
# imort fungsi label encoder ke label encoder 1
label_encoder_1 = LabelEncoder()
```

```python
# mengubah label species menjadi numeric
iris_labels = label_encoder_1.fit_transform(iris_df.Species)
```

```python
# append kolom target ke dataframe
iris_df['target'] = iris_labels

# iris_df['target']: Ini adalah cara untuk menambahkankolom di dalam DataFrame
# =: Operator penugasan ini akan menetapkan nilai dari iris_labels ke
# - dalam kolom 'target' di iris_df.
```
