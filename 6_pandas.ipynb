## **Manipulating DataFrame**

```python
# membuat data random baru untuk data frame
new_kolom = np.random.randint(1,5,(150,1))

# parameter randint(angka terkecil, terbesar(baris,kolom))
```

```python
# Menambah kolom pada dataframe
# iris_df['class'] = iris_datasets.target
iris_df['Type'] = new_kolom

# iris_df['kolom baru'] = pilih kolom baru yang ingin kamu masukkan berdasarkan dataset yang diambil
# - lihat pada feature name dataset/target
# iris_datasets.target = ini merupakan tempat data kolom baru yang diamboil
# iris_df['Iris-setosa'] = iris_datasets.target
```

```python
iris_df.head()
```

```python
# cara menghapus baris pada DataFrame
# perubahan ini hanya sementara dan tidak permanen
iris_df.drop(index=0, axis=0)

# index=0 aritnya yaitu baris pertama dimulai dari 0, index digunakan untuk memilih baris yang dihapus
# axis=0 artinya 0 menandakan baris dan 1 menandakan kolom.
# - jadi untuk menghilangkan baris menggunakan axis 0
```

```python
# cara menghapus kolom pada DataFrame
# perubahan ini hanya sementara dan tidak permanen
iris_df.drop(columns='Jenis', axis=1)

# columns='nama kolom' = memilih kolom yang ingin dihapus pada DataFrame
# axis=1 yaitu nilai 1 pada axis menandakan kolom sementara 0 menandakan baris
# - jadi untuk menghilangkan kolom menggunakan axis=1
```

```python
# Cara menghilangkan kolom atau baris secara permanan yaitu dengan membuat variabel baru untuk datasetnya
# new_iris merupakan variabel baru yang menyimpan hasil penghapusan dataset kolom sebelumnya

new_iris = iris_df.drop(columns=['class','Iris-setosa','Jenis'], axis=1)
```

```python
new_iris.head()
```

```python
# cara melihat value pada baris DataFrame
new_iris.loc[1]

# fungsi .loc[index] = masukkan nilai index baris yang ingin dilihat valuenya
```

```python
# cara melihat value pada kolom DataFrame
print(new_iris.iloc[:,0]) # melihat kolom pertama
print(new_iris.iloc[:,-1]) # melihat kolom Kelima, jika -1 artinya dimulai dari akhir kolom

# .iloc[:,0] = artinya sama dengan [baris,kolom]
# jadi tanda titik dua : yaitu mengartikan seluruh baris
# 0 artinya index pada kolom DataFrame atau index pertama
```
