# **Matplotlib**


*   Digunakan untuk membuat plot grafik



```python
# import library Matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Membuat data arrray menggunakan numpy
# linspace = parameternya yaitu angka dimulai dari 0, berhenting di angka 10,
# - dan total data arraynya yaitu 150
x = np.linspace(0, 10, 150)
y = np.sin(x) # nilai sin artinya
z = np.cos(x)
```

```python
# Membuat grafik plot sin(y)
plt.plot(x, y) # ini digunakan untuk membuat grafik plot
plt.show() # digunakan untuk menampilkan atau running plot

# jika tidak memilih jenis plotnya maka secara default akan dibuat grafik garis
```

# **Membuat Bar Chart**

Penjelasan kode dibawah:


*   plt.figure(): Fungsi ini membuat sebuah figure kosong yang menjadi wadah bagi grafik yang akan digambar.
*   fig.add_axes([0, 0, 1, 1]): Fungsi ini menambahkan axes (sumbu) pada figure yang telah dibuat sebelumnya. Parameter [0, 0, 1, 1] adalah koordinat dan ukuran sumbu dalam sistem koordinat figure:

*   bahasa adalah nilai x label dan people adalah nilai y label
*   ax.bar(bahasa, people): Fungsi ini membuat grafik batang (bar chart) pada objek ax (axes) yang telah dibuat.





```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
bahasa = ['Indonesia','Melayu','Jawa','Minang','English']
people = [100, 150, 50, 65, 85]
ax.bar(bahasa, people)
plt.xlabel('BAHASA')
plt.ylabel('JUMLAH')
plt.show()
```

# **Scatter Plot**

penjelasan kode dibawah
1. np.linspace(0, 10, 50): Fungsi linspace menghasilkan 50 titik data yang terdistribusi secara merata dari 0 hingga 10. Variabel x berisi array dengan 50 nilai yang tersebar merata dalam rentang ini, yang akan digunakan sebagai sumbu x pada grafik.
2. y = np.sin(x): Fungsi sin(x) menghitung nilai sinus untuk setiap elemen dalam array x. Hasilnya adalah array y yang berisi nilai sinus dari setiap elemen x.
3. z = np.cos(x): Fungsi cos(x) menghitung nilai kosinus untuk setiap elemen dalam array x. Hasilnya adalah array z yang berisi nilai kosinus dari setiap elemen x.

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x, y, color='g') # untuk grafik scatter 1 berwarna hijau, kode g
ax.scatter(x, z, color='r') # scatter 2 berwarna merah karena kode r
plt.show
```

# 3D Scatter Plot

Penjelasan kode dibawah:
1. plt.axes(projection='3d'): Fungsi ini membuat objek axes (sumbu) dengan proyeksi 3D, yang memungkinkan pembuatan grafik 3 dimensi.
2. 15 * np.random.random(100): Fungsi np.random.random(100) menghasilkan 100 nilai acak dalam rentang [0, 1]. Dengan mengalikannya dengan 15, Anda mendapatkan 100 angka acak antara 0 dan 15.
3. x = np.sin(z): Fungsi np.sin(z) menghitung nilai sinus untuk setiap elemen dalam array z. Hasilnya adalah array x yang berisi nilai sinus dari elemen-elemen dalam z.
4. y = np.cos(z): Fungsi np.cos(z) menghitung nilai kosinus untuk setiap elemen dalam array z
5. ax.scatter(x, y, z, c=z, cmap='Greens'): Fungsi scatter() membuat grafik sebar (scatter plot) 3D dengan data x, y, dan z:
  - x, y, z adalah data yang akan diplotkan pada sumbu x, y, dan z pada grafik 3D.
- c=z menentukan warna titik berdasarkan nilai pada z. Artinya, semakin besar nilai z, semakin gelap warna titiknya.
- cmap='Greens' menentukan colormap yang digunakan untuk mewarnai titik-titik pada grafik. Dalam hal ini, colormap 'Greens' akan memberikan warna hijau untuk titik-titik, di mana nilai z yang lebih besar akan memiliki warna yang lebih gelap.

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 15 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Greens')
plt.show()
```
