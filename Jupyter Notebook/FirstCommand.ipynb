# Visualisasi Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat data sebagai dictionary
data = {
    'Nama': ['Alice', 'Bob', 'Charlie'],
    'Usia': [25, 30, 35],
    'Kota': ['Jakarta', 'Bandung', 'Surabaya']
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Membuat bar plot untuk kolom 'Usia'
plt.figure(figsize=(8, 5))
plt.bar(df['Nama'], df['Usia'], color='skyblue')
plt.xlabel('Nama')
plt.ylabel('Usia')
plt.title('Usia per Nama')
plt.show()
```
